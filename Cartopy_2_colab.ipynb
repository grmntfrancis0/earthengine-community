```
import ee
```

```
ee.Authenticate()
ee.Initialize(project = "ee-grmntfrancis0",
opt_url='https://earthengine-highvolume.googleapis.com')
```

```
import geemap
```

```
map = geemap.Map()
map
```

```
roi = map.draw_last_feature.geometry()
roi
```

```
pdsi = (ee.ImageCollection("IDAHO_EPSCOR/TERRACLIMATE")
.filterDate("2005", "2026")
.select("pdsi")
.mean()

)
pdsi
```

```
import xarray as xr
```

```
ds = xr.open_dataset(
    pdsi,
    engine = "ee",
    crs = "EPSG:4326",
    geometry = roi,
    scale = 1
)
ds
```

```
ds = ds.squeeze("time").drop_vars("time") * 0.01
ds
```

```
!pip install cartopy
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```
plt.figure()
ax = plt.axes(projection = ccrs.EqualEarth())

ds.pdsi.plot(
    x = "lon", y = "lat", transform = ccrs.PlateCarree(),
robust = True, cmap = "RdBu_r", levels = 10,
cbar_kwargs = {
    "orientation":"horizontal",
    "pad":0.07, "shrink":0.7
}

)

ax.coastlines(resolution = "110m")
ax.set_global()
ax.gridlines(draw_labels = True, alpha = 0.5)
ax.set_title("Global Drought Map (2005-2025)", fontsize = 20)
plt.savefig("drough_map.png", dpi = 361, bbox_inches = "tight")

```
